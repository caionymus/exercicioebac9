```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

dados_sinasc = pd.read_csv('sinasc_RO_2019.csv')
```

```python
nome = dados_sinasc.columns
print (nome)
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
plt.figure(figsize=(8, 6))  
sns.boxplot(x='IDADEMAE', data=dados_sinasc, orient='h')
plt.xlabel('Idade da Mãe')
plt.title('Boxplot da Idade da Mãe - SINASC 2019')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
plt.figure(figsize=(8, 6))  
sns.boxplot(x='PESO', data=dados_sinasc, orient='h')
plt.xlabel('Peso do Bebê')
plt.title('Boxplot da Peso do Bebê - SINASC 2019')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
idade_mae = dados_sinasc['IDADEMAE']
plt.figure(figsize=(8, 6))
plt.hist(idade_mae, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência')
plt.title('Histograma da Idade da Mãe - SINASC 2019')
plt.grid(True)
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
peso_bebe = dados_sinasc['PESO']
plt.figure(figsize=(8, 6))
plt.hist(peso_bebe, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Peso do Bebê')
plt.ylabel('Frequência')
plt.title('Histograma da Peso do Bebê - SINASC 2019')
plt.grid(True)
plt.show()
```
